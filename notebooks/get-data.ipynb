# Transformación de los Datos...

```python
aspirantes_mcd = pd.DataFrame()

for year in range(2020,2024):
    df = pd.read_excel(f'..\\data\\aspirantes_{year}.xlsx')
    
    aspirantes_mcd = pd.concat([aspirantes_mcd,df])

aspirantes_mcd
```

```python
date_birth = list(aspirantes_mcd['Fecha Nacimiento'])
```

```python
age = []
for date in date_birth:
    age.append(int(date[-3:-1]))
```

```python
aspirantes_mcd['Edad'] = age

aspirantes_mcd=aspirantes_mcd.drop(columns='Fecha Nacimiento')
```

```python
status = list(aspirantes_mcd['Estatus Proceso'])

aceptado = [True if stat == '[7] ACEPTADO' else False for stat in status]

aspirantes_mcd['Aceptado'] = aceptado

aspirantes_mcd = aspirantes_mcd.drop(columns='Estatus Proceso')
```

```python
nacional = list(aspirantes_mcd['Nacionalidad'])

nacional = [True if x == 'MEXICANO' else False for x in nacional]

aspirantes_mcd['Nacionalidad'] = nacional
```

```python
aspirantes_mcd
```

```python
aspirantes_mcd['Fecha Pre-registro'] = pd.to_datetime(aspirantes_mcd['Fecha Pre-registro'])
aspirantes_mcd['Licenciatura: Fecha Titulación'] = pd.to_datetime(aspirantes_mcd['Licenciatura: Fecha Titulación'])
```

```python
aspirantes_mcd['Experiencia'] = aspirantes_mcd['Fecha Pre-registro'].dt.year-aspirantes_mcd['Licenciatura: Fecha Titulación'].dt.year
```

```python
aspirantes_mcd
```

```python
#Eliminando columnas
aspirantes_mcd = aspirantes_mcd.drop(columns = ['Religión', 'Num Hijos', 'Pais', 'Estado', 'Fecha Pre-registro', 'Licenciatura: Fecha Titulación', 'aspirantePosgradoId'])
```

# Chequeando la calidad de los datos antes de crear el Tidy...

```python
# Voy a leer los datos del Fichero que ya estaba creado porque esta libreta no corre completa...
aspirantes_mcd = pd.read_excel('aspirantes-mcd.xlsx')
aspirantes_mcd = aspirantes_mcd.drop("Unnamed: 0", axis = 1)

aspirantes_mcd
```

```python
# Total de nulos... por columnas...
aspirantes_mcd.isnull().sum()
```

```python
# Encontrar filas con valores nulos para cada columna
filas_con_nulos = aspirantes_mcd.isnull().any(axis=1)

# Imprimir las filas con valores nulos
print("Filas con valores nulos:")
print(aspirantes_mcd[filas_con_nulos])
```

```python
# Se van a impotar a "0" los valores nulos de la columna experiencia... Y "Desconocida/o" en caso que las (Instituciones, Programa y Título) estén vacías...
aspirantes_mcd["Experiencia"].fillna(0, inplace=True)
aspirantes_mcd["Licenciatura: Institución"].fillna("Desconocida", inplace=True)
aspirantes_mcd["Licenciatura: Programa"].fillna("Desconocido", inplace=True)
aspirantes_mcd["Licenciatura: Tipo Titulación"].fillna("Desconocido", inplace=True)

# Verificando que no hay valores nulos...
aspirantes_mcd.isnull().sum()
```
