```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel('aspirantes-mcd.xlsx')
```

```python
n_data = data.shape[0]

female_mcd = data[data['Sexo'] == 'MUJER']
male_mcd = data[data['Sexo'] == 'HOMBRE']

agegender, ax = plt.subplots(ncols=1, nrows=2,sharex=True, sharey=True)

ax[0].hist(female_mcd['Edad'])
ax[1].hist(male_mcd['Edad'])
ax[0].set_title('Edad de las mujeres aspirantes')
ax[1].set_title('Edad de los hombres aspirantes')

plt.savefig('..\\graficas\\age.png')
plt.show()
```

```python
yes_female = female_mcd.groupby('Aceptado')['Sexo'].count()
yes_male = male_mcd.groupby('Aceptado')['Sexo'].count()

accepted = np.array(
    [list(yes_male),
     list(yes_female)]
).transpose()
yes = pd.DataFrame(columns=['Hombre','Mujer'], index= ['No Aceptado', 'Aceptado'], data = accepted)
yes = yes * n_data ** (-1)
yes.plot(kind= 'bar', title = 'Aspirantes Aceptados', rot=0)
plt.savefig('..\\graficas\\accepted.png')
```
